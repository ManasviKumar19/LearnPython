```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11) # range 0-5 will be divided into 10 intervals   step=5-0/10
print(x)
y=x**2
print(y)
```

```python
plt.figure()
plt.plot(x,y)  #x and y must have same dimensions
plt.show()

plt.figure()
plt.plot(np.array([1,2,3,4]),np.array([4,5,6,7]))  #x and y must have same dimesions
# plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(np.array([1,2,3,4]),np.array([4,5,6,7]))  #x and y must have same dimesions
# plt.show()
```

```python
plt.figure()
plt.plot(x,y,color='c')  #c,m,y,k=black,w,r,g,b
plt.show()
```

```python
plt.figure()
plt.plot(x,y,color='#D2271A')  # hex color codes can also be given
plt.show()
```

```python
plt.figure()
plt.plot(x,y,color='c')  # hex color codes can also be given
#plt.show()
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('Graph')
plt.grid()
#plt.grid(False)  #removes grid from a graph
```

```python
plt.figure()
plt.plot(x,y,color='c') 
#plt.show()
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('Graph')
plt.axis([0,5,0,20])
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.figure(figsize=(6,4),facecolor='c')
plt.plot(x,y,color='k')
plt.grid()
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.figure(figsize=(6,4),facecolor='c')
plt.plot(x,y,color='k',alpha=0.5)   #alpha is the opacity
plt.grid()
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.figure(figsize=(6,4),facecolor='c')
plt.plot(x,y,color='k',alpha=1,linestyle='-.')   #alpha is the opacity
plt.grid()
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.figure(figsize=(6,4),facecolor='c')
plt.plot(x,y,color='k',alpha=1,linestyle='-.',linewidth=10)  
plt.grid()
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.figure()
plt.plot(x,y,'.',color='k') 
plt.grid()
```

```python
x=[5,10,7]
y=[12,6,15]
x1=[4,9,6]
y1=[11,5,14]

plt.plot(x,y,color='c')
plt.plot(x,y,'.',color='k')

plt.plot(x1,y1,color='g')
plt.plot(x1,y1,'.',color='r')
```

```python
x=np.linspace(0,5,11)
y=x**2
y1=x**3
plt.figure()
plt.plot(x,y,'r-.')
plt.plot(x,y1,'c--')
```

```python
x=np.linspace(0,5,11)
y=x**2
y1=x**3
plt.figure()
plt.plot(x,y,'r-.',label='square')
plt.plot(x,y1,'c',label='cube')
plt.legend()   #left top corner wala box...pan hyala labels pahijet
#plt.legend(loc=6)  #location of legend
plt.grid(color='m',alpha=0.3)
```

```python
x=np.linspace(0,5,11)
y=x**2
y1=x**3
y2=x**4
plt.figure()
plt.scatter(x,y,marker='*',s=200)
plt.scatter(x,y1,marker='.',s=200)
plt.scatter(x,y2,marker='.',s=200)
```

```python
x=[5,10,7]
y=[12,6,15]
x1=[4,9,7]
y1=[11,5,10]
# plt.bar(x,y,color='c')
# plt.bar(x1,y1,color='k')
plt.bar(x,y,color='c',width=0.6,align='edge')   #cyan bars are aligned at the edges of x axis elements
plt.bar(x1,y1,color='k',width=0.6)
```
