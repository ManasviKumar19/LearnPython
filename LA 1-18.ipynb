# LA 12 Factorial RECURSION

```python
a=input('enter the number:')
a=int(a)
def fact1(a):
    if(a==0):
        return 1
    else:
        f=a*fact1(a-1)
        return f

if(a<0):
    print('Enter a no greater than or equal to 0')
else:
    f=fact1(a)
    print('factorial is:',f)
```

# LA 13 DECI AND HEXA

```python
i=input('enter a decimal number:')
s=len(i)
binary=''
def binary1(y):
    c=int(y)
    binary=''
    while(int(y)>1):
        if(int(y)%2==0):
            binary=str(binary)+'0'
        else:
            binary=str(binary)+'1'
        y=int(y)/2
    if(y%2!=0):
        binary=str(binary)+'1'
      
        
    return binary
b=binary1(i)
b=b[::-1]
print('The binary number is :',b)

dict1={10:'A',11:'B',12:'C',13:'D',14:'E',15:'F'}
def hexa(y):
    c=int(y)
    h=''
    while(c>1):
        rem=c%16
        c=c//16
        if 0<rem<9:
            h=h+str(rem)
        else:
            char=dict1[rem]
            h=h+char
    return h
h=hexa(i)
h=h[::-1]
print('The hexadecimal equivalent is:',h)
```

# LA 15 INVERSE OF A MATRIX

```python
import numpy as np
import sys
while(True):
    list1=[]
    list2=[]
    list1=list(map(int,input('Enter the row 1:').split()))
    print(list1)
    list2=list(map(int,input('Enter the row 2:').split()))
    print(list2)
    list3=[list1,list2]
    mat=np.array(list3)
    print(mat)

    det=list1[0]*list2[1]-list1[1]*list2[0]
    print('det is',det)

    if det==0:
        print('Inverse isnt possible')
        continue
    else:
        adj=np.zeros((2,2))
        adj[0,0]=mat[1,1]
        adj[0,1]=-mat[0,1]
        adj[1,0]=-mat[1,0]
        adj[1,1]=mat[0,0]
        print(adj)
        det=1/det
        inv=det*adj
    print(inv)
    break
```

# LA 18 Phase Shift

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math as ma

list1=np.linspace(0,10,100)
a=np.sin(list1)
b=np.sin((2*ma.pi)/3+list1)
c=np.sin((4*ma.pi)/3+list1)
plt.figure()
plt.plot(a,'r',label='phase 1')
plt.plot(b,'g',label='phase 2')
plt.plot(c,'c',label='phase 3')
plt.grid()
plt.legend(loc=9)
```
